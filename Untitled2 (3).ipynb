# Webscraping and Exploratory Data Analysis using Python

## Data Preparation

```python
# Importing necessary libraries for data manipulation, visualization, statistical analysis, and web scraping
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import statsmodels.api as sm
import seaborn as sns
import requests
from bs4 import BeautifulSoup
```

```python
# Defining the URL to scrape data from and making an HTTP GET request to fetch the webpage content
url = 'https://www.scrapethissite.com/pages/simple/'
response = requests.get(url)
```

```python
# Parsing the HTML content of the webpage and extracting all elements with the specified class
soup = BeautifulSoup(response.text, 'html.parser')
countries = soup.find_all('div', class_='col-md-4 country')
```

```python
# creating an empty list to store the extracted data
data = []
```

```python
# Looping through each country element to extract and clean the country name, capital, population, and area.
# Then appending the extracted data as a dictionary to the data list
for country in countries:
    country_name = country.find('h3', class_='country-name').text.strip()
    capital = country.find('span', class_='country-capital').text.strip()
    population = country.find('span', class_='country-population').text.strip()
    area = country.find('span', class_='country-area').text.strip()
    data.append({
        'Country': country_name,
        'Capital': capital,
        'Population': population,
        'Area': area
    })
```

```python
# Displays list of dictionaries containing country information
print(data)
```

```python
# Converting the list of dictionaries into a pandas DataFrame and printing it to display the structured data
df = pd.DataFrame(data)
print(df)
```

## Data Cleaning

```python
# Calculating and printing the number of missing values in each column of the DataFrame
null_count = df.isnull().sum()
print(null_count)
```

```python
# Identifying and printing rows in the DataFrame that are duplicates
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
```

```python
# Printing the data types of each column in the DataFrame to understand the structure of the data
print(df.dtypes)
```

```python
# Converting the 'Population' and 'Area' columns to numeric data types, coercing any errors to NaN
df['Population'] = pd.to_numeric(df['Population'], errors='coerce')
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
```

```python
# Checking for NaN values
nan_count = df['Population'].isnull().sum()
print(nan_count)
```

## Exploratory data analysis

```python
# Sorting the DataFrame by the 'Population' column in descending order,
# selecting the top 10 countries with the highest population,
# and printing the 'Country' and 'Population' columns of these top 10 countries
df_sorted = df.sort_values(by='Population', ascending=False)
top_10_countries = df_sorted.head(10)
print(top_10_countries[['Country', 'Population']])
```

```python
# Defining a formatter function to convert large numbers into billions for display on the y-axis
def billions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fB' % (x * 1e-9)
```

```python
# Apply a theme and create a bar chart
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_countries['Country'], y=top_10_countries['Population'], hue=top_10_countries['Country'], palette='viridis', dodge=False, legend=False)
plt.xlabel('Country')
plt.ylabel('Population (in billions)')
plt.title('Top 10 Countries by Population')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(FuncFormatter(billions))
```

```python
# Creating a copy of the original DataFrame and calculating the population density for each country,
# then adding it as a new column 'Population Density (per sq km)' and printing the updated DataFrame
df_dens = df.copy()
df_dens['Population Density (per sq km)'] = df_dens['Population'] * 1000 / df['Area']
print(df_dens)
```

```python
# Calculating and printing the number of NaN values in the 'Population Density (per sq km)' column
nan_count = df_dens['Population Density (per sq km)'].isnull().sum()
print(nan_count)
```

```python
# Filling the NaN value with 0 as the 0 makes sense in this context
df_dens.fillna(0, inplace=True)
```

```python
# Checking to see if it was replaced 
nan_count = df_dens['Population Density (per sq km)'].isnull().sum()
print(nan_count)
```

```python
# Sorting the DataFrame by 'Population Density (per sq km)' in descending order,
# selecting the top 10 records with the highest population density
df_dens=df_dens.sort_values(by='Population Density (per sq km)', ascending=False)
top_10_records_dens=df_dens.head(10)
```

```python
# Defining a formatter function to convert large numbers into millions for display on the y-axis
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)
```

```python
# Apply theme and create bar chart
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_records_dens['Country'], y=top_10_records_dens['Population Density (per sq km)'], hue=top_10_records_dens['Country'], palette='viridis', dodge=False, legend=False)
plt.xlabel('Country')
plt.ylabel('Population Density (per sq km)')
plt.title('Top 10 Countries with Highest Population Density')
plt.xticks(rotation=45)
plt.ylim(0, 18000000)
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))
plt.show()
```

```python
# Defining variables for regression analysis
X = df['Area']
y = df['Population']
```

```python
# Adding a constant to the independent variables
# Fitting the Ordinary Least Squares (OLS) model
# Printing the summary of the model
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Calculates the residuals from the fitted model and creates a scatter plot of residuals versus 'Area', to address model results 
residuals = model.resid
plt.scatter(df['Area'], residuals)
plt.xlabel('Area')
plt.ylabel('Residuals')
plt.title('Residuals vs Area')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
# Plot a histogram of the residuals
plt.hist(residuals, bins=30)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
# Fit a robust linear model using Huber's T norm and print the summary
rlm_model = sm.RLM(y, X, M=sm.robust.norms.HuberT()).fit()
print("\nM-estimation Summary (Huber's T):")
print(rlm_model.summary())
```
